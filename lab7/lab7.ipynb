```python
import numpy as np
import pandas as pd
```

#### Определение генератора случайных величин

```python
def generator(n):
    for i in range(n):
        rp = np.random.uniform(0, 1) ** (1 / 3)
        rq = np.random.uniform(0, 1) ** (1 / 3)
        mq = 2 * np.random.uniform(0, 1) - 1
        yield rp, rq, mq
```

#### Расчет ρ и *l*

```python
rho = lambda toss: np.sqrt(toss[0] ** 2 + toss[1] ** 2 - 2 * toss[0] * toss[1] * toss[2])
```

```python
rho_denom = lambda toss: 1 / rho(toss)
```

```python
rho_2 = lambda toss: toss[0] ** 2 + toss[1] ** 2 - 2 * toss[0] * toss[1] * toss[2]
```

```python
rho_denom2 = lambda toss: 1 / rho_2(toss)
```

```python
l = lambda toss: toss[0] * toss[2] + np.sqrt(1 - toss[0] ** 2 * (1 - toss[2] ** 2))
```

#### Определение двух методов вычисления интеграла

```python
def mmk1(tosses, n):
    sum1, sum2 = 0, 0
    for toss in tosses:
        sum1 += rho_denom(toss)
        sum2 += rho_denom2(toss)
    integral_1 = 16 * sum1 / 9 / n
    integral_2 = 16 * sum2 / 9 / n
    return integral_1, integral_2    
```

```python
def mmk2(tosses, n):
    sum1, sum2 = 0, 0
    for toss in tosses:
        sum1 += l(toss) ** 2 * toss[1] ** 3
        sum2 += l(toss)
    integral_1 = 16 * sum1 / 3 / n
    integral_2 = 16 * sum2 / 3 / n
    return integral_1, integral_2    
```

```python
res = {
    'Метод 1, f = 1/ρ': [],
    'Метод 1, f = 1/ρ²': [],
    'Метод 2, f = 1/ρ': [],
    'Метод 2, f = 1/ρ²': []
    }

res_error = {
    'Метод 1, f = 1/ρ, %': [],
    'Метод 1, f = 1/ρ², %': [],
    'Метод 2, f = 1/ρ, %': [],
    'Метод 2, f = 1/ρ², %': []
}

n = 100_000

true1 = 32 / 15
true2 = 4

for i in range(50):
    method1 = mmk1(generator(n), n)
    method2 = mmk2(generator(n), n)
    res['Метод 1, f = 1/ρ'].append(method1[0])
    res['Метод 1, f = 1/ρ²'].append(method1[1])
    res['Метод 2, f = 1/ρ'].append(method2[0])
    res['Метод 2, f = 1/ρ²'].append(method2[1])
    
    res_error['Метод 1, f = 1/ρ, %'].append(np.round((np.abs(true1 - method1[0]) / true1) * 100, 2))
    res_error['Метод 1, f = 1/ρ², %'].append(np.round((np.abs(true2 - method1[1]) / true2) * 100, 2))
    res_error['Метод 2, f = 1/ρ, %'].append(np.round((np.abs(true1 - method2[0]) / true1) * 100, 2))
    res_error['Метод 2, f = 1/ρ², %'].append(np.round((np.abs(true2 - method2[1]) / true2) * 100, 2))
    
results = pd.DataFrame(res)
results_error = pd.DataFrame(res_error)
results
```
