```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Определение заданной функции

```python
f = lambda x: -(x ** 3) + x + 2
```

#### Определение производной функции в точке

```python
def derivative(func):
    def df(x):
        EPS = 10 ** (-6)
        f_plus = func(x + EPS)
        f_minus = func(x - EPS)
        return (f_plus - f_minus) / 2 / EPS
    return df
```

#### Определение итератора для метода касательных 

```python
def tanget(f, x):
    return x - (f(x) / derivative(f)(x))
```

#### Определение итератора для упрощённого метода Ньютона

```python
def newtons_iteration(f, x, x0):
    return x - (f(x) / derivative(f)(x0))
```

#### Ввод кол-ва итераций, границ интервала поиска и выбор нулевого узла

```python
k_iterations = [3, 5, 10, 20]
a, b = [0.8, 2]
x0 = a if f(a) * derivative(derivative(f))(a) > 0 else b 
```

#### Реализация методов для разного кол-ва итераций и вывод результатов

```python
for k in k_iterations:
    x = x0
    for i in range(k):
        x = tanget(f, x)
        
    print(f'Метод касательных для {k} итераций выдал: {x = }, {f(x) = }')
    x = x0
    for i in range(k):
        x = newtons_iteration(f, x, x0)
    print(f'Упрощённый метод Ньютона для {k} итераций выдал: {x = }, {f(x) = }')
```

#### Визуализация графиков исходной функции, первой и второй производных

```python
X = np.linspace(-2, 2, num=100, endpoint=True)
y = [f(x) for x in X]
dy = [derivative(f)(x) for x in X]
ddy = [derivative(derivative(f))(x) for x in X]
fig, ax = plt.subplots()
plt.plot(X, y, label="f(x)")
plt.plot(X, dy, label="df(x)")
plt.plot(X, ddy, label="d^2(f(x))")
plt.legend(fontsize = 15)
fig.set_figwidth(10)
fig.set_figheight(10)
ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('y', fontsize=15)
ax.grid()
plt.savefig('Lab1')
plt.show()
```
